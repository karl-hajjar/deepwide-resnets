# Glorot init

```python
#step = 50
(start, num) = (10, 10)
betas = [0.25, 0.5, 1.0] #[0., 0.25, 0.5, 1.0]
all_x = torch.randn(size=(N_TRIALS, N_SAMPLES, INPUT_DIM))
n_res_grid = np.linspace(start=start, stop=N_RES, num=num, dtype=int)

ratios = dict()
th_ratios = dict()
s = time()
with torch.no_grad():
    for beta in betas:
        ratios_ = {n_res: [] for n_res in n_res_grid}
        th_ratios_ = {n_res: [] for n_res in n_res_grid}
        for n_res in n_res_grid:
            for i in range(N_TRIALS):
                alpha = n_res**(-beta)
                net = ResNet(input_dim=INPUT_DIM, width=WIDTH, activation=ACTIVATION, bias=BIAS, alpha=alpha, 
                             n_res=n_res, kind=KIND)
                x = all_x[i, :, :]

                h_0 = net.input_layer(x)
                h_L = net.residual_layers(h_0)
                
                #h_0_norm = torch.sum(h_0**2, dim=1) / net.d_model
                h_0_norm = torch.sum(h_0**2, dim=1)
                factor = (1 + alpha**2)**(n_res) - 1
#                ratios_[n_res].append(torch.mean((torch.sum((h_L-h_0)**2, dim=1) / net.d_model) /
#                                                  h_0_norm).detach().item())
                ratios_[n_res].append(torch.mean((torch.sum((h_L-h_0)**2, dim=1)) /
                                                  h_0_norm).detach().item())
                
                ratios_[n_res].append(factor)

        ratios[beta] = ratios_
        th_ratios[beta] = th_ratios_
        
e = time()
print('Time for L={:,} and N_TRIALS={:,} : {:.2f} minutes'.format(N_RES, N_TRIALS, (e - s)/60))
```

```python
plt.figure(figsize=(10, 6))
plt.yscale('log')
marker = 'o'
#plt.xscale('log')
colors = sns.color_palette()

for i, key in enumerate(ratios.keys()):
    data = pd.DataFrame(ratios[key])
    mins = data.min(axis=0).values
    maxs = data.max(axis=0).values
    means = data.mean(axis=0).values
    
    delta = 1 - 2 * key

    plt.plot(data.columns.values, means, marker=marker, label='$\\beta = {}$'.format(key), c=colors[i])
    #plt.plot(n_residuals, means, label='$\\beta = {}$'.format(key))
    plt.fill_between(data.columns.values, mins, maxs, alpha=0.2, color=colors[i])

    if delta > 0:
        plt.plot(np.arange(start, N_RES+1), np.exp((np.arange(start, N_RES+1) ** delta) -0.5), 
                 label='$\\exp(L^{1 -2\\beta})$', c=colors[i], linestyle='dashed')
    elif delta < 0:
        plt.plot(np.arange(start, N_RES+1), np.arange(start, N_RES+1)  ** (delta), 
                 label='$L^{-(2\\beta - 1)}$', c=colors[i], linestyle='dashed')
    else:  # delta = 0
        plt.plot(np.arange(start, N_RES+1), len(np.arange(start, N_RES+1)) * [np.exp(1) - 1], 
                 label='$e - 1$', c=colors[i], linestyle='dashed')

plt.xlabel('$L$ (# residual layers) ', fontsize=28)
plt.ylabel('$||h_L - h_0||^2 \, / ||h_0||^2$', fontsize=28)
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.legend(fontsize=24)
#plt.savefig(os.path.join(FIGURES_DIR, '.png'))
plt.show()
```
